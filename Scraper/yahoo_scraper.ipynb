```python
import random
# ranking function
def spearman_ranking(a):
    input_array = np.array(a)
    sorted_indices = np.argsort(-input_array)
    ranks = np.empty_like(sorted_indices)
    ranks[sorted_indices] = np.arange(1, len(input_array) + 1)
    return ranks

# spearman
# need to have ranking function implemented
# THIS IS INCOMPLETE 
# NEED TO ADD 
def spearman_coefficient(stock1, stock2):
    a = spearman_ranking(stock1)
    b = spearman_ranking(stock2)
    n = (len(a) + len(b)) / 2
    rho = 1 - (6 * np.sum((a - b)**2) / (n * (n**2 - 1)))
    #print(a)
    #print(b)
    #print(a-b)
    #print(np.sum(a-b))
    return rho

example1 = []
example2 = []

for i in range(250):
    example1.append(random.randint(1,500))
    example2.append(random.randint(500,600))

#%timeit spearman_coefficient(example1, example2)
#example1= [1,2,3,4,4,5,6,1]
#example2 = [5,14,23,54,76,104,78,1000]
example1 = [1,2,3,4,5,6,7]
example2 = [7,6,5,4,3,2,1]
print(spearman_coefficient(example1, example2))
```
